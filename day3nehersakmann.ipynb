```python
df['56'] = df['56'].astype(int)
type("56")
import plotly.io as pio
```

```python
fig = px.line(df, y="56")
fig.update_layout(template="plotly_dark")
fig.update_traces(line_color='white')
fig.update_xaxes(visible=False, fixedrange=True)
fig.update_yaxes(visible=False, fixedrange=True)
fig.add_annotation(x=815, y=-100,
            text="10 pA",
            showarrow=False,
            yshift=10)
fig.add_annotation(x=680, y=-260,
            text="250 ms",
            showarrow=False,
            yshift=10)

fig.add_shape(type="line",
    x0=600, y0=-230, x1=750, y1=-230, line_width=12)
fig.add_shape(type="line",
    x0=780, y0=-150, x1=780, y1=-50, line_width=12)
fig.show()
fig.write_image("images/figpng.png")
fig.to_image(format="png", engine="kaleido")
```
